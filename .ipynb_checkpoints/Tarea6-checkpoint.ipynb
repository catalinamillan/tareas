## Curva de peso promedio (Splines)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
x_array=[0,6,10,13,17,20,28] # Tiempo en dias
y1_array=[6.67,17.33,4.67,37.33,30.10,29.31,28.74] #Muesta 1
y2_array=[6.67,16.11,18.89,15.,10.56,9.44,8.89]  #Muestra 2
```

```python
cs1= CubicSpline(x_array,y1_array)
cs2= CubicSpline(x_array,y2_array)
```

```python
xs = np.linspace(0,30,1000)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_array,y2_array,'*', label='Data2')
plt.plot(xs,cs2(xs),label= 'Spline2')
plt.plot(x_array,y1_array,'*', label='Data1')
plt.plot(xs,cs1(xs),label= 'Spline1')
plt.show()
```

## Interp Hermite

```python
t_list=[0,3,5,8,13]
x_list=[0,225,383,623,993]
dxdt_list=[75,77,80,74,72]
```

```python
Q = np.zeros((2*len(t_list),2*len(t_list)))
Z = np.zeros(2*len(t_list))

for i in range(0,len(t_list)):
    Z[2*i]      = t_list[i]
    Z[2*i+1]    = t_list[i]
    Q[2*i][0]   = x_list[i]
    Q[2*i+1][0] = x_list[i]
    Q[2*i+1][1] = dxdt_list[i]
    if i!=0:
        Q[2*i][1] = (Q[2*i][0]-Q[2*i-1][0]) / (Z[2*i]-Z[2*i-1])
        
for i in range(2,len(Z)):
    for j in range(2,i+1):
        Q[i][j] = (Q[i][j-1]-Q[i-1][j-1]) / (Z[i]-Z[i-j])


def PolyHermite(x):
    sum = 0
    for i in range(1,len(Z)):
        prod = 1.0
        for j in range(0,i):
            prod = prod*(x-Z[j])
        sum=sum+Q[i][i]*prod
    H = Q[0][0]+sum 
    return H
```

```python
print('La posición en t=10s es %.5e m' %(PolyHermite(10)))
```
