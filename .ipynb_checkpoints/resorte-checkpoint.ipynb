# Oscilador Armónico
## Ejercicio 1 

Gráfica de la posición, velocidad, aceleración y energías de un oscilador

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Parámetros
m=0.5 #Kg
t=0 #s
xi=2 #m
k=1.5 #kg/s^2 (constante elástica)
w=(k/m)**.5 #rad/s
dt=0.1 #s
```

```python
X=[]  #Posición
V=[]  #Velocidad
ac=[] #Aceleración
T=[]  #Tiempo
Ep=[] #Energía potencial
Ek=[] #Energía cinética
ET=[] #Energía total
```

```python
while t<15:
    x=xi*np.cos(w*t)
    X.append(x)
   
    v=-xi*w*np.sin(w*t)
    V.append(v)
    
    a=-xi*w**2*np.cos(w*t)
    ac.append(a)
    
    Ek.append(.5*m*v**2.)
    Ep.append(.5*k*x**2.)
    
    t=t+dt
    T.append(t)
    
for i in range(len(Ek)):
    ET.append(Ek[i]+Ep[i])
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(T,X,'.',label='Posicion')
ax.plot(T,V, '+',label='Velocidad')
ax.plot(T,ac,'-.', label='Aceleracion') 
ax.grid() 
ax.set_ylabel(r"Amplitud(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14) 
plt.title("Posicion, velocidad y Aceleracion", fontsize=16)
plt.legend(loc='upper right')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(T,Ep,label='E-potencial')
ax.plot(T,Ek,label='E-cinetica')
ax.plot(T,ET,'-.',label='E-total')
ax.grid() 
ax.set_ylabel(r"Amplitud(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14) 
plt.title("Energia cinetica, potencial y total", fontsize=16)
plt.legend(loc='lower right')
plt.show()
```
